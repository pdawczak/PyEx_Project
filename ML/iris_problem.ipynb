```python
import pandas as pd
```

```python
dataset = pd.read_csv("./Iris.csv", index_col=0)
```

```python
dataset
```

```python
X = dataset[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
y = dataset['Species']
```

```python
X
```

```python
y
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
```

```python
model.fit(X, y)
```

```python
model.predict([[5.1, 3.5, 1.4, 0.2]])
```

```python
model.predict([[5.9, 3.0, 5.1, 1.8]])
```

```python
import pickle
```

```python
pickle.dump(model, open('iris_model.pickle', 'wb'))
```
